```python
def count_zero_positions(filename: str) -> int:
    # Dial always ranges from 0 to 99
    MOD = 100

    # Starting position
    pos = 50

    # Counter for how many times we land on 0
    zero_count = 0

    with open(filename, "r") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            direction = line[0]      # 'L' or 'R'
            distance = int(line[1:]) # number after direction

            if direction == 'L':
                pos = (pos - distance) % MOD
            else:  # direction == 'R'
                pos = (pos + distance) % MOD

            if pos == 0:
                zero_count += 1

    return zero_count


if __name__ == "__main__":
    filename = "day1_input.txt"   # <-- your puzzle input file
    result = count_zero_positions(filename)
    print("Password =", result)
```
